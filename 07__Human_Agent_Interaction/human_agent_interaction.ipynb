## Module Theory: Steerability & A2UI

**What is Steerability?**
Steerability guarantees that an autonomous agent can be paused, corrected, or completely halted by a human operator mid-execution.

**Agent-to-User Interface (A2UI):**
When an agent encounters a low-confidence decision, it must block its execution graph and surface the context to a human. This turns pure autonomy into **Human-in-the-Loop (HITL)** architecture.

Key Mechanisms:
*   Confidence Thresholds
*   Execution Pausing (Interrupts)
*   State Resurrection
