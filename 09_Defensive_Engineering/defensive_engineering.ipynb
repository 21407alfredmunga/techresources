## Module Theory: Model Armor & Trust Chains

**Adversarial Reality**
As soon as an agent is allowed to execute tools (write database, send email), prompt injection becomes Remote Code Execution (RCE).

*   **Model Armor:** A specialized middleware that sits *before* the main agent, specifically trained to classify if a user prompt contains jailbreak attempts or harmful intent.
*   **Trust Chain Verification:** Every action an agent takes must be crypto-graphically signed or verified against a strict whitelist of allowed operations. If an agent tries to invoke a tool off-whitelist, it is terminated.
