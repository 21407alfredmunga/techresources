## Module Theory: Smart Routing & Token Budgeting

**The Cost of Autonomy**
Agents loop. An unoptimized agent might burn thousands of tokens polling for status or making trivial choices.

**Smart Model Routing (SMR):**
Don't use GPT-4/Gemini Pro for every task. Route trivial categorization and extraction to smaller models (GPT-3.5/Gemini Flash) and reserve heavy logical reasoning for frontier models.

**Token Budgeting:**
Set absolute limits on agent loops to prevent infinite loops from draining resources.
