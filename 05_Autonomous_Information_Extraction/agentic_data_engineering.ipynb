## Module Theory: Entity Resolution & Knowledge Graphs

**What is Entity Resolution?**
It is the systematic extraction of discrete entities (e.g., people, organizations, locations) and their relationships from unstructured text.

*   **Nodes:** The entities themselves.
*   **Edges:** The relationships connecting them.
*   **Graph Linking:** Resolving multiple mentions of an entity across documents to a single definitive node.

By feeding a Knowledge Graph to an agent, you ground its reasoning in a deterministic structure rather than relying solely on parametric memory.
