## Module Theory: Hierarchical vs. Diamond Patterns

**Multi-Agent Execution Patterns**

1.  **Hierarchical Routing:** A prominent 'Manager' agent assesses a task and delegates it down to specialized 'Worker' agents.
2.  **Diamond Execution (Scatter-Gather):** A task is broadcasted to multiple parallel agents, and their discrete outputs are gathered and synthesized by a final 'Synthesizer' agent.

In this module, we focus on dynamic routing based on task complexity.
